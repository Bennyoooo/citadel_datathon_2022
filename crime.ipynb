```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
from google.colab import drive
drive.mount('/content/drive')
at = 'drive/MyDrive/datathon/Copy of Austin Crime Reports.csv'
```

Austin

```python
at_ci = pd.read_csv(at)
at_ci
```

```python
at_ci.columns
```

```python
at_ci['Occurred Date'] =  pd.to_datetime(at_ci['Occurred Date'], format="%m/%d/%Y")
```

```python
at_ci_selected = at_ci[(at_ci['Occurred Date'] > "12/31/2019") & (at_ci['Occurred Date'] < "01/01/2022")]
```

```python
at_by_day = at_ci_selected.groupby('Occurred Date').size().reset_index(name='count')
at_by_day
```

```python
at_by_day.to_csv("at_by_day.csv", sep='\t')
```

```python
at_by_month = at_ci_selected.groupby(pd.Grouper(key='Occurred Date', freq='M')).size().reset_index(name='count')
at_by_month
```

```python
at_by_month.to_csv("at_by_month.csv", sep='\t')
```

```python
at_cd = at_ci_selected[at_ci_selected['Council District'].notna()]
at_by_cdday = at_cd.groupby(['Occurred Date', 'Council District']).size().reset_index(name='count')
at_by_cdday
```

```python
at_by_cdday.to_csv("at_by_cdday.csv", sep='\t')
```

DC

```python
dc2020 = 'drive/MyDrive/datathon/Copy of DC Crime Incidents 2020.csv'
dc2021 = 'drive/MyDrive/datathon/Copy of DC Crime Incidents 2021.csv'
dc_2020 = pd.read_csv(dc2020)
dc_2021 = pd.read_csv(dc2021)
```

```python
dc_ci = pd.concat([dc_2020, dc_2021])
dc_ci
```

```python
dc_ci['REPORT_DAT'] =  pd.to_datetime(dc_ci['REPORT_DAT'], format="%Y/%m/%d %H:%M:%S")
```

```python
dc_ci['report_date'] = dc_ci['REPORT_DAT'].dt.date
```

```python
dc_ci
```

```python
dc_ci_selected = dc_ci[(dc_ci['REPORT_DAT'] > "2019-12-31") & (dc_ci['REPORT_DAT'] < "2022-01-01")]
dc_by_month = dc_ci_selected.groupby(pd.Grouper(key='REPORT_DAT', freq='M')).size().reset_index(name='count')
dc_by_month
```

```python
dc_by_day = dc_ci_selected.groupby(pd.Grouper(key='REPORT_DAT', freq='D')).size().reset_index(name='count')
dc_by_day
```

```python
dc_by_wardday = dc_ci_selected.groupby(['report_date', 'WARD']).size().reset_index(name='count')
dc_by_wardday
```

```python
dc_by_month.to_csv("dc_by_month.csv", sep='\t')
dc_by_day.to_csv("dc_by_day.csv", sep='\t')
dc_by_wardday.to_csv("dc_by_wardday.csv", sep='\t')
```

NYC

```python
ny2021 = 'drive/MyDrive/datathon/Copy of NYPD Calls for Service 2021.csv'
ny_2021 = pd.read_csv(ny2021)
ny_2021
```

```python
ny_2021 = ny_2021[ny_2021['CIP_JOBS']=='Critical']
```

```python
ny2020 = 'drive/MyDrive/datathon/Copy of NYPD Calls for Service 2020.csv'
ny_2020 = pd.read_csv(ny2020)
```

```python
ny_2020 = ny_2020.drop('OBJECTID', 1)
```

```python
ny_2020
```

```python
ny_2020 = ny_2020[ny_2020['CIP_JOBS']=='Critical']
```

```python
ny_ci = pd.concat([ny_2020, ny_2021])
ny_ci
```

```python
ny_ci['INCIDENT_DATE'] =  pd.to_datetime(ny_ci['INCIDENT_DATE'], format="%m/%d/%Y")
```

```python
ny_by_month = ny_ci.groupby(pd.Grouper(key='INCIDENT_DATE', freq='M')).size().reset_index(name='count')
ny_by_month
```

```python
ny_by_day = ny_ci.groupby('INCIDENT_DATE').size().reset_index(name='count')
ny_by_day
```

```python
ny_by_borday = ny_ci.groupby(['INCIDENT_DATE','BORO_NM']).size().reset_index(name='count')
ny_by_borday
```

```python
ny_by_month.to_csv("ny_by_month.csv", sep='\t')
ny_by_day.to_csv("ny_by_day.csv", sep='\t')
ny_by_borday.to_csv("ny_by_borday.csv", sep='\t')
```
