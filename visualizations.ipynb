```python
def call_freq_df(df, time_range, complaint_type):
  
  """Takes in a dataframe and a complaint type. Returns a dataframe with dates and the total 
     number of calls related to the given complaint that were made on each date."""
    
  subset = df[df['Complaint Type']==complaint_type]
    
  call_freqs_by_date = {}
    
  for dt in subset['Created Date']:
    if time_range == 'day':
      date = datetime.strptime(dt,'%m/%d/%Y %I:%M:%S %p').date()
    elif time_range == 'month':
      date = datetime.strftime(datetime.strptime(dt,'%m/%d/%Y %I:%M:%S %p'), '%m/%Y')
    if date not in call_freqs_by_date: 
      call_freqs_by_date[date] = 1
    else:
      call_freqs_by_date[date] += 1
    
  freq_df = pd.DataFrame.from_dict(call_freqs_by_date, orient='index', columns=['{}'.format(complaint_type)])
    
  return freq_df
```

```python
nyc_311_reopen_call_freq = call_freq_df(nyc_311_df, 'day', 'NonCompliance with Phased Reopening')
nyc_311_reopen_call_freq.plot()
```

```python
nyc_311_noise1_call_freq = call_freq_df(nyc_311_df, 'month', 'Noise - Residential')
nyc_311_noise2_call_freq = call_freq_df(nyc_311_df, 'month', 'Noise - Street/Sidewalk')
nyc_311_noise3_call_freq = call_freq_df(nyc_311_df, 'month', 'Noise - Vehicle')
nyc_311_noise4_call_freq = call_freq_df(nyc_311_df, 'month', 'Noise - Commercial')
nyc_311_noise5_call_freq = call_freq_df(nyc_311_df, 'month', 'Noise')
```

```python
plt.rcParams["figure.figsize"] = (18,7)
plt.plot(nyc_311_noise1_call_freq, label='Noise - Residential')
plt.plot(nyc_311_noise2_call_freq, label='Noise - Street/Sidewalk')
plt.plot(nyc_311_noise3_call_freq, label='Noise - Vehicle')
plt.plot(nyc_311_noise4_call_freq, label='Noise - Commercial')
plt.plot(nyc_311_noise5_call_freq, label='Noise - Undefined')
plt.legend()
```

```python
def plot_calls(month_df, month):
  fig = px.density_mapbox(month_df, lat=month_df['Latitude'], lon=month_df['Longitude'], radius=2, animation_frame=month_df['day'],
                          hover_name=None, hover_data=['Complaint Type','Community Board'], 
                          width=550, height=550, color_continuous_scale=px.colors.sequential.Inferno)
  fig.update_layout(mapbox_style="carto-positron", mapbox_zoom=8.5, mapbox_center = {"lat": 40.6885, "lon": -73.93211},)                    
  fig.layout.coloraxis.showscale = False    
                      
  fig.update_layout(title={
        'text': 'Call Volume in {}'.format(month),
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}),

  font=dict(family="silom",
              size=14, color="#58508d")

  fig.update_layout(transition = {'duration': 10})

  fig.show()
```

```python
plot_calls(nyc_311_df_aug[:100000], 'October')
```
